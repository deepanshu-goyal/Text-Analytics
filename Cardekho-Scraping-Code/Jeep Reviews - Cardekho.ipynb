```python
#Import Libraries required for web scraping
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
#This code will open the chrome browser and then open the cardekho website
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.cardekho.com/jeep/compass/user-reviews')
```

```python
#This code will find the url link of the 'full review' on the first page. 
url=[]
for data in driver.find_elements_by_class_name('readReviewHolder'):
    url.append(data.find_element_by_tag_name('a').get_attribute('href'))
```

```python
#This code will find the url link of the 'full review' on all the 6 pages.
for i in range(2,7,1):
    cook_url = 'https://www.cardekho.com/jeep/compass/user-reviews/'+ str(i)
    driver.get(cook_url)
    for data in driver.find_elements_by_class_name('readReviewHolder'):
        url.append(data.find_element_by_tag_name('a').get_attribute('href'))
```

```python
url
```

```python
#This code will fetch review information from all the review URLs. 
car_review=[]
brand = 'Jeep Compass'
for each_url in url:
    print(each_url)
    driver.get(each_url)
    time.sleep(1)
    title = driver.find_element_by_tag_name('h1').text
    print(title)
    data = driver.find_element_by_class_name('readReviewBox')
    review = data.find_element_by_class_name('contentheight').text
    print(review)
    author_name = data.find_element_by_class_name('name').text[3:]
    print(author_name)
    review_date = data.find_element_by_class_name('date').text[4:16]
    print(review_date)
    views = data.find_element_by_class_name('date').text[19:]
    print(views)
    like_count = data.find_elements_by_class_name('hover')[0].text
    print(like_count)
    dislike_count = data.find_elements_by_class_name('hover')[1].text
    print(dislike_count)
    rating=0
    try:
        for i in range(1,6,1):
            data.find_element_by_class_name('stars.stars'+str(i)+'.icon-star-full-fill ')
            rating+=1
    except EC.NoSuchElementException:
        pass
    print(rating)
    car_review.append({'Brand':brand,'Title':title,'Review':review,'Rating':rating,
                       'Author Name': author_name, 'Review Date':review_date,
                       'Views':views,'Like Count':like_count,'Dislike Count':dislike_count,'URL':each_url,                    
                      })
```

```python
car_review
```

```python
#Save customer review in csv format
final_data = pd.DataFrame(car_review, columns=['Brand','Title','Review','Rating','Author Name','Review Date',
                                               'Views','Like Count','Dislike Count','URL'])
final_data.to_csv('JeepCompass_reviews.csv')
```
