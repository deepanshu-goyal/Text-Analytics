```python
### Setup Logging to run the entire script to understand step by step progress of the script

def setup_log():
    logger = logging.getLogger(__name__)
    ch = logging.StreamHandler(sys.stdout)
    ch.setLevel(os.environ.get("LOGLEVEL", "INFO"))
    formatter = logging.Formatter('[%(asctime)s] [%(filename)s:%(lineno)d:%(funcName)s()] [%(levelname)s] - %(message)s')
    ch.setFormatter(formatter)
    logger.addHandler(ch)
    return logger
```

```python
### Opening the Chrome in incognito mode to avoid blocking of ip

option = webdriver.ChromeOptions()
option.add_argument("â€” incognito")
```

```python
### Launches headless Chrome Browser

browser = webdriver.Chrome(executable_path="/usr/bin/chromedriver", chrome_options=option)
```

```python
### No Pages to be scrapped

pages = 10
```

```python
### This block helps create A dataframe conistiting of the Brand Name and the review of that car brand with the date of 
### the review and the ratings as well

total = []
cars = ["kia seltos","MG Hector","Jeep Compass"]
urls = ["https://www.cardekho.com/kia/seltos/user-reviews","https://www.cardekho.com/mg/hector/user-reviews",
       "https://www.cardekho.com/jeep/compass/user-reviews"]
for car,url_car in zip(cars,urls):
    try:
        for page in range(2,pages):
            print("page",page)
            url = url_car+"/"+str(page)

            browser.get(url)
            reviews_list = len(browser.find_elements_by_class_name("reviewList"))

            for reviews in range(reviews_list):
                review = browser.find_elements_by_class_name("readReviewBox")
                for context in review:
                    car_name = car
                    context_text = context.find_element_by_class_name('contentheight').text
                    date = context.find_element_by_class_name('date').text
                    pattern = "\w+\s\d{2},\s\d{4}"
                    string = "On: Sep 16, 2019 | 1114 Views"
                    text = re.search(pattern,string)
                    count = 0
                    for i in range(1,6):
                        try:
                            context.find_element_by_css_selector("span.stars.stars"+str(i)+".icon-star-full-fill")
                            count +=1
                        except Exception as e:
                            context.find_element_by_css_selector("span.stars.stars"+str(i)+".icon-star-full-empty")
                            count -=1
                    new = ((car_name,context_text,text.group(0),np.abs(count)))
                    print(new)
                    total.append(new)
            df1 = pd.DataFrame(total,columns=['Brand','Review','Review_date','Rating_stars'])
    except Exception as e:
        pass
        #df.to_csv('quoted.csv')
```

```python
### Converts the dataframe into csv file

df1.to_csv("User_review_data.csv",index=False)
```
